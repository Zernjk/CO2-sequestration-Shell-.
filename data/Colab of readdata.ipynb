```python
df_G = pd.read_csv(file_path+'spe1_data_3/G_centroids.csv', header=None)
print(df_G)
```

```python
df_s = list()
df_p = list()
df_ps = list()
for i in range(1,121):
    df_s.append(pd.read_csv(file_path+'spe1_data_3/s_'+str(i)+'.csv', header=None))
    df_p.append(pd.read_csv(file_path+'spe1_data_3/p_'+str(i)+'.csv', header=None))

    df_ps.append(pd.concat([df_p[i-1], df_s[i-1]], axis=1))
```

```python
df_rock_perm = pd.read_csv(file_path+'spe1_data_3/rock_perm.csv', header=None)
print(df_rock_perm)
```

```python
df_rock_poro = pd.read_csv(file_path+'spe1_data_3/rock_poro.csv', header=None)
print(df_rock_poro)
```

```python
# construct dataset with pressure and satura+tion
with open(file_path+"spe1_data_3/pressure.pkl", 'wb') as f1:
    pickle.dump(df_p, f1) 

with open(file_path+"spe1_data_3/saturation.pkl", 'wb') as f2:
    pickle.dump(df_s, f2) 
```

```python
with open(file_path+"spe1_data_3/df_ps_lst.pkl", 'wb') as f3:
    pickle.dump(df_ps, f3) 
```
